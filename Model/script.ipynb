```python
import pandas as pd
import numpy as np
import datetime
```

```python
data = pd.read_csv('./BSACTData.csv')
```

```python
data.head()
```

```python
data.index = pd.to_datetime(data.Timestamp, format="%d-%m-%Y %H:%M")
```

```python
data.columns
```

```python
sig1, sig2, sig3 = data['median-confirmation-time'].dropna().std(), data['transaction-fees'].dropna().std(), data['n-transactions-per-block'].dropna().std()
```

```python
allDates = np.unique(data.index.date)

for indx, date in enumerate(allDates):
    if date == datetime.date(2020, 5, 3):
        continue
    dd = data.loc[date + datetime.timedelta(days=1), :]
    mu1 = dd['median-confirmation-time']
    mu2 = dd['transaction-fees']
    mu3 = dd['n-transactions-per-block']
    mask = data.index.date == date
    data.loc[mask, 'median-confirmation-time'] = mu1 + np.random.randn(sum(mask))*sig1
    data.loc[mask, 'transaction-fees'] = mu2 + np.random.randn(sum(mask))*sig2
    data.loc[mask, 'n-transactions-per-block'] = mu3 + np.random.randn(sum(mask))*sig3
    print(indx+1, "/", len(allDates))
    # break
```
